# Tilfeldig? 

Litt system i kaoset? Teiknar tilfeldige punkt basert p√• terning. 
Inspirert av Aleksander Seland og Visual Math sin [video](https://www.facebook.com/Visual.maths2000/videos/488065012954166/?extid=NS-UNK-UNK-UNK-IOS_GK0T-GK1C)

### Algoritmen:
1) Teikn ein likesida trekant ABC
2) Velg eit tilfeldig startpunkt inne i trekanten
3) Trill ein terning
- Viss talet blir 1 eller 2 flytt punktet halvvegs til A
- Viss talet blir 3 eller 4 flytt punktet halvvegs til B
- Viss talet blir 5 eller 6 flytt punktet halvvegs til C
4) Velg det nye punktet som nytt startpunkt
5) Gjenta punkt 3 og 4

```python
import matplotlib.pyplot as plt
import numpy as np
import random

def kaos_og_terning(N):
    plt.figure(figsize = (7,7))
    s = 10 #sidelengd i trekanten
    
    A = np.array([0, 0])
    B = np.array([s, 0])
    C = np.array([s/2, np.sqrt(s**2-(s/2)**2)])

    # startpunkt
    P = np.array([2, 2])

    plt.plot(A[0], A[1], 'k.')
    plt.plot(B[0], B[1], 'k.')
    plt.plot(C[0], C[1], 'k.')
    plt.plot(P[0], P[1], 'k.')

    for i in range(1,N):
        t = random.randint(1, 6)
    
        if t == 1 or t == 2:
            P = P + 0.5*(A-P)
        elif t == 3 or t == 4:
            P = P + 0.5*(B-P)
        else:
            P = P + 0.5*(C-P)
            
        plt.plot(P[0], P[1], 'k.')
    plt.xlim(-1, 11)
    plt.ylim(-1, C[1] + 1)
    plt.show()
```

```python
kaos_og_terning(1)
```

```python
kaos_og_terning(10)
```

```python
kaos_og_terning(100)
```

```python
kaos_og_terning(1000)
```

```python
kaos_og_terning(10000)
```

```python
kaos_og_terning(100000)
```

## Kvadrat

Pr√∏ver det samme med eit kvadrat.

```python
def kaos_og_terning_kvadrat(N):
    plt.figure(figsize = (7,7))
    s = 10 #sidelengd i trekanten
    
    A = np.array([0, 0])
    B = np.array([s, 0])
    C = np.array([s, s])
    D = np.array([0, s])
    
    # startpunkt
    P = np.array([2, 2])

    plt.plot(A[0], A[1], 'k.')
    plt.plot(B[0], B[1], 'k.')
    plt.plot(C[0], C[1], 'k.')
    plt.plot(D[0], D[1], 'k.')
    plt.plot(P[0], P[1], 'k.')

    forrige = 0
    
    for i in range(1,N):
        t = random.randint(1, 4)
        
        if t != forrige:               # lar ikkje hj√∏rnet vera det samme som f√∏rre runde
            if t == 1:
                P = P + 0.5*(A-P)
            elif t == 2:
                P = P + 0.5*(B-P)
            elif t == 3:
                P = P + 0.5*(C-P)
            else:
                P = P + 0.5*(D-P)
                
            plt.plot(P[0], P[1], 'k.')
            
        forrige = t
    plt.xlim(-1, 11)
    plt.ylim(-1, C[1] + 1)
    plt.show()
```

```python
kaos_og_terning_kvadrat(10000)
```
